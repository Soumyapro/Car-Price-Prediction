### Cleaning Data

```python
df = df[df['year'].str.isnumeric()]
df['year'] = df['year'].astype(int)
```

```python
df = df[df['Price'] != "Ask For Price"].copy()
df['Price'] = df['Price'].str.replace(",", "").astype(int)
```

```python
df['kms_driven'] = df['kms_driven'].str.split(" ").str.get(0).str.replace(",","")
```

```python
df = df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven'] = df['kms_driven'].astype(int)
```

```python
df.isna().sum()
```

```python
df = df[~df['fuel_type'].isna()]
```

```python
df['name'] = df['name'].str.split(' ').str.slice(0,3).str.join(' ')
print(df['name'])
```

```python
df = df.reset_index(drop=True)
print(df.head(10))
```

```python
df.describe()
```

```python
df = df[df['Price']<6e6].reset_index(drop=True)
```

```python
df.describe()
```

### Training and Testing Data

```python
X = df.drop(columns='Price')
y = df['Price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.2)
```

```python
one_hot_encoder = OneHotEncoder()
one_hot_encoder.fit(X[['name','company','fuel_type']])
```

### Creating an OneHotEncoder object to contain all the possible categories

```python
column_trans = make_column_transformer((OneHotEncoder(categories=one_hot_encoder.categories_),['name','company','fuel_type']), remainder='passthrough', force_int_remainder_cols=False)
```

```python
lr = LinearRegression()
```

### Creating a Pipeline

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores = []
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
print(scores[np.argmax(scores)])
```

### Best Model found at Random State 433

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=433)
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
```

```python
r2 = r2_score(y_test, y_pred)
print(f"R² Score: {r2:.4f}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Car Prices")
plt.show()
```
